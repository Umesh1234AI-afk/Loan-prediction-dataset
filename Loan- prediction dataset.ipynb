# Handling Missing Value:-

```python
df.isnull().sum()
```

```python
The dataset contains a small number of missing values in important columns such as Gender, Dependents, Self_Employed, LoanAmount, Loan_Amount_Term,

and Credit_History.

The highest missing values are observed in Credit_History and Self_Employed, which may impact the reliability of financial analysis.
```

```python
# Fill categorical missing values with mode

df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])

df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])

df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])


# Fill numerical missing values with median

df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())

df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
```

# Insight:-

Missing values in categorical columns were handled using mode imputation, while numerical columns were treated using median imputation.
    
This preprocessing step improved data consistency and ensured that the dataset became complete and ready for further exploratory data analysis.
